```python
#modelling a bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f'{amount} is deposited in the bank new bank balance is {self.balance}')

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficent funds!")
        else:
            self.balance-=amount
            print(f'{amount} is withdrawm from the bank new bank balance is {self.balance}')

    def get_balance(self):
        print(f'Your current bank balance is:{self.balance}')


account=BankAccount('soham',1000000)
account.deposit(500000)
account.withdraw(500000)
account.get_balance()
```

```python
class Counter:
    count=0#it is a class variable

    def __init__(self):
        Counter.count+=1#Counter.count is used to deal with the class variable count

    @classmethod#this is used in defining method in class refernce which are not affiliated with object instances
    def get_count(cls):#for object instance we use self but for class refernce we use "cls"
        return cls.count

c1=Counter()
c2=Counter()
c3=Counter()
c4=Counter()
print(c1.get_count())
```

```python
#polymorphism
class Shape:
    def area(self):
        return "The shape of area"

class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width*self.height
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius*self.radius



def print_Area(shape):
    print(f"The area of the shape is{shape.area()}")


s1=Shape()

r1=Rectangle(3,2)
c1=Circle(5)
print_Area(r1)
print_Area(s1)
print_Area(c1)


```

```python
from abc import ABC ,abstractmethod

class Vechile(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vechile):
    def start_engine(self):
        return "Car engine started"
    
class MotorCycle(Vechile):
    def start_engine(self):
        return "MotorCycle engine started"
    
class BiCycle(Vechile):
    def start_engine(self):
        pass
        
        

    

def start_vechile(vechile):
    print(vechile.start_engine())

b1=BiCycle()
c1=Car()
m1=MotorCycle()
start_vechile(c1)
start_vechile(m1)
```

```python
#Encapsulation with getter and setter method
class Person:
    def __init__(self,name,age):
        self.__name=name#double underscore means that it is private variblae and can be accessed within the class and its derived classes only
        self.__age=age

    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def set_name(self,name):
        self.__name=name
    
    def set_age(self,age):
        self.__age=age


p1=Person("Soham",34)
print(p1.get_name())
print(p1.get_age())
p1.set_age(45)
p1.set_name("Shweta")
print(p1.get_name())
print(p1.get_age())
        
```

```python
#ABSTRACTION

from abc import ABC,abstractmethod

class Vechile(ABC):

    def drive(self):
        print("i am driving a vechile")

    @abstractmethod
    def start_engine(self):
        pass

class car(Vechile):
    def start_engine(self):
        print("The car engine is started")

class motorcyle(Vechile):
    pass

def operate_engine(vechile):
    vechile.start_engine()
    vechile.drive()


c1=car()
m1=motorcyle()
operate_engine(c1)
operate_engine(m1)
```
