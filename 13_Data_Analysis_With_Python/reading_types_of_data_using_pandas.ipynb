READING DATA FROM JSON USING PANDAS

```python
import pandas as pd
from io import StringIO

data = '''
{
    "employee_name":"James",
    "email":"james@gmail.com",
    "job_profile":[
        {
            "title1":"Team lead",
            "title2":"Sr.Developer"
        }
    ]
}
'''

df = pd.read_json(StringIO(data))

print(df)

```

```python
df
```

```python
df.to_json()
```

HERE ORIENT IS MEANS Which direction should Pandas arrange the data in JSON?"

```python
df.to_json(orient="index")
```

```python
df.to_json(orient="records")
```

READING DATA FROM WIKIPEDIA(HTML) USING PANDAS 

```python
!pip install requests
```

HERE WE INSTALL REQUESTS BECAUSE 
Very Simple Understanding

When you open a website in Chrome:

Browser → sends request → Website server

The server then sends back:

HTML
JSON
images
data

requests lets Python behave like a browser.

```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent":"Mozilla/5.0"
}

response = requests.get(url, headers=headers)

df = pd.read_html(response.text, match="Country", header=0)

print(df[0])
```

```python
df[0]
```
