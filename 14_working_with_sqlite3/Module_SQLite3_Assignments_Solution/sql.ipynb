SQL AND SQLite

SQL (Structured Query Language) is a standard language for managing and manipulating relational databases.

SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems.

In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3  
```

```python
##connect to sqllite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
#create a table
cursor.execute(
    '''
    Create Table If Not Exists employees(
                    id Integer Primary Key,
                    name Text Not Null,
                    age Integer,
                    dept Text)
    '''
)

#commit changes
connection.commit()
```

```python
cursor.execute(
    '''
    select * from employees'''
)
```

```python
##insert data into sqlite3 table
cursor.execute(
    '''
    insert into employees(name,age,dept) values('krish',32,'data scientist')
    '''
)
cursor.execute(
    '''
    insert into employees(name,age,dept) values('bob',25,'engineering')
    '''
)
cursor.execute(
    '''
    insert into employees(name,age,dept) values('charlie',35,'finanace')
    '''
)

#commmit changes
connection.commit()
```

```python
cursor.execute(
    '''
    select * from employees'''
)
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
#update the table
cursor.execute(
    '''
    update employees set age=34 where name='krish'
    '''
)

connection.commit()
```

```python
cursor.execute(
    '''
    select * from employees'''
)
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
#delete from table 
cursor.execute(
    '''
    delete from employees where name='krish'
    '''
)
connection.commit()
```

```python
cursor.execute(
    '''
    select * from employees'''
)
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
##working with sales data
connection=sqlite3.connect('sales_db')
cursor=connection.cursor()

cursor.execute(
    '''
    create table if not exists sales(
    id integer primary key,
    date text not null,
    product text not null,
    sales integer,
    region text)
    '''
)

sales_data=[
    ('2023-01-01','Product1',100,'North'),
    ('2023-01-02','Product2',150,'South'),
('2023-01-03','Product3',200,'East'),
('2023-01-04','Product4',250,'West'),
('2023-01-05','Product5',300,'North'),
('2023-01-06','Product6',180,'South')
]


cursor.executemany(
    '''
    insert into sales(date,product,sales,region) values(?,?,?,?)''',sales_data
)
```

```python
connection.commit()
```

```python
cursor.execute(
    '''
    select * from sales'''
)
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
##close the connection
connection.close()
```
