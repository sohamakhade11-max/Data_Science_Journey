LOGGIN WITH PYTHON 

Logging in Python is a built-in mechanism for recording events, messages, warnings, errors, and debugging information that occur while a program is running. It is provided by Python's logging module and helps developers monitor, debug, and maintain applications.

```python
import logging
```

Level	Numeric Value	Purpose

DEBUG	10	Detailed information for debugging

INFO	20	General information about program execution

WARNING	30	Something unexpected happened, but program continues

ERROR	40	A serious problem occurred in a specific operation

CRITICAL	50	Very serious error; application may stop

```python
#configure login
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)
```

```python

logging.debug('this is an debug message')
logging.info('this is info message')
logging.warning('this is a warning message')
logging.error('this is an error message')
logging.critical('this is an critical message')
```

Easy-to-Remember Definition

Think of logging as a diary for your program.

Instead of printing messages on the screen using print(), logging keeps a record of what your program is doing, what went wrong, and when it happened.

For example:

User logged in → recorded
File opened successfully → recorded
Database connection failed → recorded
Application crashed → recorded
