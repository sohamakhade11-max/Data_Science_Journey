```python
with open('example.txt',mode='r') as file:
    content=file.read()
    print(content)
```

```python
with open('example.txt',mode='r') as file:
    for row in file:
        print(row.strip())#.strip() removes the newline character
```

```python
with open('example.txt',mode='a',newline='') as file:
    file.write("\nAPPEND footballl")

with open('example.txt',mode='r') as file:
    content=file.read()
    print(content)

file.close()
```

```python
lines=['first line\n','second line\n','third line\n']

with open('example.txt','a') as file:
    file.writelines(lines)

with open('example.txt',mode='r') as file:
    content=file.read()
    print(content)

file.close()

```

```python
import datetime

def log_msg(msg,filename='activity.log'):
    timestamp=datetime.datetime.now().isoformat()
    with open(filename,'a') as file:
        file.write(f'[{timestamp}] {msg}\n')

log_msg('i am spider man')
```
